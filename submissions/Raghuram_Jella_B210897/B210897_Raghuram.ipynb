## 5.2 *args and **kwargs

````python
# *args — accepts any number of positional arguments
def add_all(*args):
    print(f"Received: {args}")   # it's a TUPLE
    return sum(args)

print(add_all(1, 2, 3))          # 6
print(add_all(10, 20, 30, 40))   # 100
````

````python
# **kwargs — accepts any number of keyword arguments
def print_info(**kwargs):
    print(f"Received: {kwargs}")   # it's a DICT
    for key, value in kwargs.items():
        print(f"  {key}: {value}")

print_info(name="Ravi", age=25, city="Hyderabad")
````

## 5.3 Lambda Functions (🔥 Used in pandas, ML pipelines)

````python
# Regular function
def square(x):
    return x ** 2

# Lambda — same thing, one line
square = lambda x: x ** 2
print(square(5))     # 25

# Lambda with multiple arguments
add = lambda a, b: a + b
print(add(3, 7))     # 10
````

````python
# Real-world: sorting with lambda
students = [("Ravi", 85), ("Priya", 92), ("Amit", 78)]
students.sort(key=lambda x: x[1], reverse=True)
print(students)   # Sorted by marks descending

# Real-world: map + lambda
numbers = [1, 2, 3, 4, 5]
doubled = list(map(lambda x: x * 2, numbers))
print(doubled)    # [2, 4, 6, 8, 10]

# Real-world: filter + lambda
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)      # [2, 4]
````

## 7.2 Your First Class — The Custom Data Type

````python
# WHY do we need this?
# Because list, tuple, dict, set — none of them can represent
# a STUDENT with data AND behavior bundled together.

class Student:
    
    def __init__(self, name, roll_no, marks):
        self.name = name           # attribute (data)
        self.roll_no = roll_no     # attribute (data)
        self.marks = marks         # attribute (data)
    
    def average(self):             # method (behavior)
        return sum(self.marks) / len(self.marks)
    
    def grade(self):               # method (behavior)
        avg = self.average()
        if avg >= 90: return 'A+'
        elif avg >= 80: return 'A'
        elif avg >= 70: return 'B'
        elif avg >= 60: return 'C'
        else: return 'F'


# Create OBJECTS (instances of the class)
s1 = Student("Ravi", 101, [85, 90, 78])
s2 = Student("Priya", 102, [92, 95, 88])

# Now data AND behavior live TOGETHER
print(f"{s1.name}: Average = {s1.average():.1f}, Grade = {s1.grade()}")
print(f"{s2.name}: Average = {s2.average():.1f}, Grade = {s2.grade()}")

# Check the type!
print(type(s1))    # <class '__main__.Student'> — YOUR custom class!
````

````python
# Compare: list/dict way vs class way

# ❌ OLD WAY — data and logic are separate
student_dict = {"name": "Ravi", "marks": [85, 90, 78]}
avg = sum(student_dict["marks"]) / len(student_dict["marks"])  # logic is outside

# ✅ NEW WAY — data and logic are bundled
s1 = Student("Ravi", 101, [85, 90, 78])
avg = s1.average()    # much cleaner! The object knows how to compute its own average
````

## 7.3 More Practical Example: BankAccount

````python
class BankAccount:
    
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
        self.transactions = []
    
    def deposit(self, amount):
        self.balance += amount
        self.transactions.append(f"Deposited ₹{amount}")
        print(f"₹{amount} deposited. Balance: ₹{self.balance}")
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds!")
            return
        self.balance -= amount
        self.transactions.append(f"Withdrew ₹{amount}")
        print(f"₹{amount} withdrawn. Balance: ₹{self.balance}")
    
    def statement(self):
        print(f"\n--- Statement for {self.owner} ---")
        for t in self.transactions:
            print(f"  {t}")
        print(f"  Current Balance: ₹{self.balance}")


# Use it!
acc = BankAccount("Ravi", 1000)
acc.deposit(5000)
acc.withdraw(2000)
acc.withdraw(10000)    # Insufficient funds!
acc.deposit(3000)
acc.statement()
````

## 7.4 OOP Concepts

### Encapsulation — Protecting data

````python
class Employee:
    
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary     # __ makes it PRIVATE
    
    def get_salary(self):          # controlled access
        return self.__salary
    
    def give_raise(self, amount):
        if amount > 0:
            self.__salary += amount
            print(f"{self.name} got a raise! New salary: ₹{self.__salary}")


emp = Employee("Ravi", 50000)
print(emp.name)              # Works fine
print(emp.get_salary())      # 50000 — through method

# Can't access directly:
try:
    print(emp.__salary)      # This will CRASH — it's private!
except AttributeError as e:
    print(f"Error: {e}")

emp.give_raise(10000)
````

### Inheritance — Reusing code

````python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return "Some sound"

class Dog(Animal):           # Dog INHERITS from Animal
    def speak(self):
        return "Woof!"

class Cat(Animal):           # Cat INHERITS from Animal
    def speak(self):
        return "Meow!"


animals = [Dog("Buddy"), Cat("Whiskers"), Dog("Max")]

for animal in animals:
    print(f"{animal.name} says {animal.speak()}")

# Buddy says Woof!
# Whiskers says Meow!
# Max says Woof!

print(isinstance(animals[0], Dog))      # True
print(isinstance(animals[0], Animal))   # True — Dog IS an Animal
````

### Polymorphism — Same method, different behavior

````python
# Polymorphism = "many forms"
# The SAME method name behaves differently for different classes

class Circle:
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14159 * self.radius ** 2

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height


shapes = [Circle(5), Rectangle(4, 6), Circle(3)]
for shape in shapes:
    print(f"{type(shape).__name__}: area = {shape.area():.2f}")

# Same method name .area() — different behavior!
````

### `__str__` and `__repr__` — Making objects printable

````python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    def __str__(self):
        return f"{self.name} — ₹{self.price}"
    
    def __repr__(self):
        return f"Product('{self.name}', {self.price})"


p = Product("Laptop", 65000)
print(p)           # Calls __str__  → Laptop — ₹65000
print(repr(p))     # Calls __repr__ → Product('Laptop', 65000)
````

````python
# ═══════════════════════════════════════════════════════
#  THE GRAND REVEAL
# ═══════════════════════════════════════════════════════

# Remember when you did this?
my_list = [1, 2, 3]
my_list.append(4)          # You called a METHOD on an OBJECT of class list

my_string = "hello"
my_string.upper()          # You called a METHOD on an OBJECT of class str

my_dict = {"a": 1}
my_dict.keys()             # You called a METHOD on an OBJECT of class dict

# Someone at Python HQ wrote:
#   class list:
#       def append(self, item): ...
#       def sort(self): ...
#
#   class str:
#       def upper(self): ...
#       def split(self): ...

# And NOW you can do the same!
class MyList:
    def __init__(self):
        self.data = []
    def add(self, item):
        self.data.append(item)
    def show(self):
        print(self.data)

ml = MyList()
ml.add(10)
ml.add(20)
ml.show()     # [10, 20]

print(type(ml))            # <class '__main__.MyList'>
print(type(my_list))       # <class 'list'>

# See? Your class is JUST LIKE Python's built-in classes.
# You're a class creator now. 🎉
````

**Q1.** Write a function `is_palindrome(text)` that returns True if a string reads the same forward and backward (ignore case).

````python
# YOUR CODE HERE

# Test:
# is_palindrome("Racecar") → True
# is_palindrome("Python") → False
def is_palindrome(str):
    if str.lower() == str[::-1].lower() :
        return True
    else :
        return False

print(is_palindrome("Racecar"))
print(is_palindrome("Python"))
````

**Q1.** Create a `Rectangle` class with `width` and `height` attributes, and methods `area()`, `perimeter()`, and `is_square()`.

````python
# YOUR CODE HERE

# Test:
# r = Rectangle(5, 5)
# r.area() → 25
# r.is_square() → True
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

    def is_square(self):
        return self.width == self.height
    
r1 = Rectangle(10, 8)
r2 = Rectangle(7, 7)

print("Area:", r1.area())
print("Perimeter:", r1.perimeter())
print("Is Square:", r1.is_square())

print("Is Square (r2):", r2.is_square())
````

**Q2.** Create a `BankAccount` class with `deposit()`, `withdraw()`, and `get_balance()` methods. The balance should be private (use `__balance`). Prevent negative withdrawals.

````python
# YOUR CODE HERE
class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance   

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
        else:
            print("Deposit amount must be positive")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be positive")
        elif amount > self.__balance:
            print("Insufficient balance")
        else:
            self.__balance -= amount

    def get_balance(self):
        return self.__balance

acc = BankAccount(1000)

acc.deposit(500)
acc.withdraw(300)
acc.withdraw(2000)   

print("Balance:", acc.get_balance())
````

**Q3.** Create a base class `Shape` with a method `area()`. Create child classes `Circle` and `Triangle` that inherit from `Shape` and implement their own `area()`. Demonstrate polymorphism.

````python
# YOUR CODE HERE
class Shape:
    def area(self):
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius


class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height


shapes = [Circle(5), Triangle(4, 6)]

for shape in shapes:
    print("Area:", shape.area())
````

**Q4.** Create a `Student` class with `__str__` method. Create 5 student objects, store them in a list, and sort by average marks using a lambda.

````python
# YOUR CODE HERE
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks   

    def average(self):
        return sum(self.marks) / len(self.marks)

    def __str__(self):
        return f"{self.name} - Avg: {self.average():.2f}"

students = [
    Student("Raghu", [85, 90, 88]),
    Student("Ram", [92, 95, 91]),
    Student("Vijay", [78, 82, 80]),
    Student("Rashmika", [88, 84, 90]),
    Student("Karan", [70, 75, 72])
]

students_sorted = sorted(students, key=lambda s: s.average(), reverse=True)


for s in students_sorted:
    print(s)
````

**Q6.** Create a `Library` class that manages a collection of books. It should support:
- `add_book(title, author)`
- `remove_book(title)`
- `search_by_author(author)` → returns list of books
- `show_all()` → prints all books

Use a list of dictionaries internally.

````python
# YOUR CODE HERE
class Library:
    def __init__(self):
        self.books = []   

    def add_book(self, title, author):
        self.books.append({"title": title, "author": author})

    def remove_book(self, title):
        for book in self.books:
            if book["title"].lower() == title.lower():
                self.books.remove(book)
                print(f"Removed: {title}")
                return
        print("Book not found")

    def search_by_author(self, author):
        result = [book for book in self.books if book["author"].lower() == author.lower()]
        return result

    def show_all(self):
        if not self.books:
            print("Library is empty")
        else:
            for book in self.books:
                print(f"{book['title']} by {book['author']}")


lib = Library()

lib.add_book("Cheppura", "Raju")
lib.add_book("AI chaduvu", "Abhi")
lib.add_book("Remedial top", "Suman")

print("\nAll Books:")
lib.show_all()

print("\nBooks by Ravi:")
print(lib.search_by_author("Ravi"))

lib.remove_book("Cheppura")

print("\nAfter Removal:")
lib.show_all()
````

## 🎯 Bonus Challenge (Optional)

Create a `Calculator` class that:
- Stores a running result (starts at 0)
- Has methods: `add()`, `subtract()`, `multiply()`, `divide()`, `reset()`, `get_result()`
- Supports chaining: `calc.add(5).multiply(3).subtract(2).get_result()` → 13
- Has a `__str__` that shows the current result

Hint: return `self` from each method to enable chaining.

````python
# YOUR CODE HERE
class Calculator():
    def __init__(self):
       self.res = 0
    def add(self, value):
        self.res += value
        return self
    def subtract(self, value):
        self.res -= value
        return self
    def multiply(self, value):
        self.res *= value
        return self
    def divide(self, value):
        if value == 0:
            print("cannot divide by zero")
        else:
            self.res /= value
        return self
    def reset(self):
        self.res = 0
        return self
    def get_result(self):
        return self.res
    def __str__(self):
        return f"Current Result: {self.res}"
    
calc = Calculator()
result = calc.add(5).multiply(3).subtract(2).get_result()
print(result)
````
